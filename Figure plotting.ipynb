```python
import numpy as np
from qutip import *
import matplotlib.pyplot as plt
import matplotlib as mpl
from scipy.signal import find_peaks
```

```python
points = 2000
deltas = {}
for k in range(len(els)):
    for J in cavcoup:
        if J == g/10: maxima = 50
        if J == g: maxima = 100
        if J == g*10: maxima = 200
        for spread in delta:
            we = spread
            ind = delta.index(spread)
#             we = [spread]*C*E
            H = a[0].dag()* a[0] * wa[0] + a[1].dag()* a[1] * wa[1]
            for i in range(len(s)):
                H = H+ s[i].dag() * s[i] * we[i]

            for i in range(E):
                H = H + g*(s[i].dag() * a[0] + a[0].dag() * s[i])

            for i in range(E):
                H = H + g*(s[i+2].dag() * a[1] + a[1].dag() * s[i+2])

            H = H - J*(a[0].dag()*a[1] + a[1].dag()*a[0])

            c_ops = []
            for i in range(len(a)):
                c_ops.append((np.sqrt(kappa) * a[i]))
            for i in range(len(s)):
                c_ops.append(np.sqrt(gamma)*s[i])
            c_ops.append(np.sqrt(power)*a[0].dag())

            wlist = np.linspace(-maxima, maxima, points)
            eigens = H.eigenstates()
            eigenvects = []
            for j in range(len(eigens[0])):
                state = []
                for i in range(len(eigens[0])):
                    state.append(np.real(eigens[1][j][i][0][0]))
                eigenvects.append(state)
            
            if els[k] == 'cav1':
                spec = spectrum(H, wlist, c_ops, a[0].dag(), a[0])
#                 spec = spec - min(spec)
#                 spec = spec/max(spec)
                fname = f+'/cav1/spread'+ lbls[ind] + '_J' + str(J) + '.txt'
                np.savetxt(fname, np.transpose([wlist, spec]))
                fname = f+'/cav1/spread'+ lbls[ind] + '_J' + str(J) + '_eigenvals.txt'
                np.savetxt(fname, np.transpose([eigens[0]]))
                fname = f+'/cav1/spread'+ lbls[ind] + '_J' + str(J) + '_eigenvects.txt'
                np.savetxt(fname, np.transpose(eigenvects))

            if els[k] == 'cav2':
                spec = spectrum(H, wlist, c_ops, a[1].dag(), a[1])
#                 spec = spec - min(spec)
#                 spec = spec/max(spec)
                fname = f+'/cav2/spread'+ lbls[ind] + '_J' + str(J) + '.txt'
                np.savetxt(fname, np.transpose([wlist, spec]))
                fname = f+'/cav2/spread'+ lbls[ind] + '_J' + str(J) + '_eigenvals.txt'
                np.savetxt(fname, np.transpose([eigens[0]]))
                fname = f+'/cav2/spread'+ lbls[ind] + '_J' + str(J) + '_eigenvects.txt'
                np.savetxt(fname, np.transpose(eigenvects))

            if els[k] == 'em1':
                spec = spectrum(H, wlist, c_ops, s[k-2].dag(), s[k-2])
#                 spec = spec - min(spec)
#                 spec = spec/max(spec)
                fname = f+'/em1/spread'+ lbls[ind] + '_J' + str(J) + '.txt'
                np.savetxt(fname, np.transpose([wlist, spec]))
                fname = f+'/em1/spread'+ lbls[ind] + '_J' + str(J) + '_eigenvals.txt'
                np.savetxt(fname, np.transpose([eigens[0]]))
                fname = f+'/em1/spread'+ lbls[ind] + '_J' + str(J) + '_eigenvects.txt'
                np.savetxt(fname, np.transpose(eigenvects))

            if els[k] == 'em2':
                spec = spectrum(H, wlist, c_ops, s[k-2].dag(), s[k-2])
#                 spec = spec - min(spec)
#                 spec = spec/max(spec)
                fname = f+'/em2/spread'+ lbls[ind] + '_J' + str(J) + '.txt'
                np.savetxt(fname, np.transpose([wlist, spec]))
                fname = f+'/em2/spread'+ lbls[ind] + '_J' + str(J) + '_eigenvals.txt'
                np.savetxt(fname, np.transpose([eigens[0]]))
                fname = f+'/em2/spread'+ lbls[ind] + '_J' + str(J) + '_eigenvects.txt'
                np.savetxt(fname, np.transpose(eigenvects))

            if els[k] == 'em3':
                spec = spectrum(H, wlist, c_ops, s[k-2].dag(), s[k-2])
#                 spec = spec - min(spec)
#                 spec = spec/max(spec)
                fname = f+'/em3/spread'+ lbls[ind] + '_J' + str(J) + '.txt'
                np.savetxt(fname, np.transpose([wlist, spec]))
                fname = f+'/em3/spread'+ lbls[ind] + '_J' + str(J) + '_eigenvals.txt'
                np.savetxt(fname, np.transpose([eigens[0]]))
                fname = f+'/em3/spread'+ lbls[ind] + '_J' + str(J) + '_eigenvects.txt'
                np.savetxt(fname, np.transpose(eigenvects))

            if els[k] == 'em4':
                spec = spectrum(H, wlist, c_ops, s[k-2].dag(), s[k-2])
#                 spec = spec - min(spec)
#                 spec = spec/max(spec)
                fname = f+'/em4/spread'+ lbls[ind] + '_J' + str(J) + '.txt'
                np.savetxt(fname, np.transpose([wlist, spec]))
                fname = f+'/em4/spread'+ lbls[ind] + '_J' + str(J) + '_eigenvals.txt'
                np.savetxt(fname, np.transpose([eigens[0]]))
                fname = f+'/em4/spread'+ lbls[ind] + '_J' + str(J) + '_eigenvects.txt'
                np.savetxt(fname, np.transpose(eigenvects))
            
```

```python

g = 5
f = ''
cavcoup = [g/10, g, 10*g]
# cavcoup = [g/10]
els = ['cav1', 'cav2', 'em1', 'em2', 'em3', 'em4']

ratio = [.1, 1, 10]
spreadlbls = ['epsilon', 'gquarter', 'g']
lbls1 = ['Emitter 2', 'Emitter 1', 'Cavity 1']
lbls2 = ['Emitter 4', 'Emitter 3', 'Cavity 2']
colors = ['orange', '#ff7f0e', '#1f77b4']
C = 2
E = 2
peaks = []
fig, ax = plt.subplots(3,3, sharex = True, sharey = True)
for i in range(len(spreadlbls)):
    P = []
    for J in cavcoup:
        left_peaks = []
        l = []
        maxs = []
        for k in range(C):
            fname = f+'cav' + str(k+1) + '/spread' + spreadlbls[i] + '_J' + str(J) +'.txt'
            wlist, spec = np.loadtxt(fname, unpack = True)
            l.append(spec)
            maxs.append(max(spec))
        for k in range(E*2):
            fname = f+'em' + str(k+1) + '/spread' + spreadlbls[i] + '_J' + str(J) +'.txt'
            wlist, spec = np.loadtxt(fname, unpack = True)
            l.append(spec)
            maxs.append(max(spec))

        for m in range(len(l)):
            left_peaks.append(l[m][find_peaks(l[m], width = 2)[0][0]])
            
        cav1 = [left_peaks[3] + left_peaks[2] + left_peaks[0], left_peaks[2] + left_peaks[0], left_peaks[0]]/(left_peaks[0] + left_peaks[2] + left_peaks[3] + left_peaks[1] + left_peaks[4] + left_peaks[5])
        cav2 = [left_peaks[5] + left_peaks[4] + left_peaks[1], left_peaks[4] + left_peaks[1], left_peaks[1]]/ (left_peaks[1] + left_peaks[4] + left_peaks[5] + left_peaks[0] + left_peaks[2] + left_peaks[3])
        P.append(left_peaks)
        
        for n in range(len(cav1)):
            ax[cavcoup.index(J)][i].bar(0, cav1[n], label = lbls1[n], color = colors[n])
            ax[cavcoup.index(J)][i].bar(1, cav2[n], label = lbls2[n], color = colors[n])
            
            
            ax[cavcoup.index(J)][i].set_xticks([0, 1])
#             ax[cavcoup.index(J)][i].set_yticks([])
#             fig.legend()
    peaks.append(P)
    
fig.suptitle('Ratios of Transmission Spectra Peaks')
ax[2][0].set_xlabel('Node')
ax[2][1].set_xlabel('Node')
ax[2][2].set_xlabel('Node')

ax[0][0].set_ylabel('$J/g = 0.1$ \n Probability')
ax[1][0].set_ylabel('$J/g = 1$ \n Probability')
ax[2][0].set_ylabel('$J/g = 10$ \n Probability')

fig.savefig(f+'ratios.png')
```
